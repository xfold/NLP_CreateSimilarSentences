## Forcing a deeper search, translating to 5 consecutive languages before going back to english

As we see here consecutive translations may result in drifts from the original sentence meaning. Nevertheless the results are also interesting :).
Also, setting variable `caseSensitive=False` makes sure uppercase letters do not get mixed in the results. 

n : number of similar sentences 5
seqdepth : number of random translations before going back to original language
maxseqdepth : breaking threshold, to stop executions from running forever if different sentences can't be found
seed : random seed
caseSensitive : consider lower/upper sentences as different sentences

```python
sentence = 'The house is on fire, call the firemen!'
sc.CreateSimilarSentences(sentence,
                    n = 5, 
                    seqdepth=5,
                    maxseqdepth=6,
                    seed=42,
                    caseSensitive=False)
```
